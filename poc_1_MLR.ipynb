**# X,Y split:**

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```python
X
Y
```

**# Train_Test_Split:**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=7)
```

```python
X_train.shape
Y_train.shape
```

**# Model_Buliting:**

```python
from sklearn.linear_model import LinearRegression

```

```python
std=LinearRegression()
```

```python
std.fit(X_train,Y_train)
```

# Model_Evalution/Model_Prediction:

```python
Y_predict=std.predict(X_test)
```
